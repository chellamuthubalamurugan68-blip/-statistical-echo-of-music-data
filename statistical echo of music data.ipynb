```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Problem01
#Import the dataset as a DataFrame and drop the first column.
url='https://www.dropbox.com/s/2hg67jin2n852mz/top50spotify.csv?dl=1'
data=pd.read_csv(url)

#drop the first column
data=data.iloc[:,1:]
data
```

```python
#Save it as 'top50.csv'.
a=data.to_csv("top50.csv")
a
```

```python
#Find the average Energy and Length of first 10 songs.
b=data.head(10)
average_energy=b["Energy"].mean()
average_length=b["Length."].mean()
print("The Average of Energy and Length of first 10 songs is",average_energy,"and",average_length)
```

```python
#Find the total length of songs, group by genre from top to bottom.
total_length_of_songs=data.groupby('Genre')['Length.'].sum().sort_values(ascending=False).to_frame()
total_length_of_songs
```

```python
#Print the artist name with the most number of tracks in one genre.
most_tracks=data.groupby(['Artist.Name','Genre']).size().sort_values(ascending=False)
most_tracks
```

```python
#Print the data of the tracks created by the artist from the previous question.
artist_name=most_tracks.index[0][0]
genre=most_tracks.index[0][1]
tracks_by_artist=data[(data['Artist.Name']==artist_name)&(data['Genre']==genre)]
tracks_by_artist
```

```python
#Convert the created series into DataFrame and replace the null values with zeroes
data1={'English':{'Sam':60,'Jackson':74,'Ahree':85},
       'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
       'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
       'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
       'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
data1=pd.DataFrame(data1).fillna(0)
data1
```

```python
#Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.
transpose_data=data1.transpose()
average_data=transpose_data.mean(axis=1).to_frame(name='average')
new_data=transpose_data.reset_index().merge(average_data.reset_index(),on='index',how='left')
new_data
```

```python
#Problem04
#import the dataset
url = 'https://www.dropbox.com/s/idnul34dfo5cnke/cereal.csv?dl=1'
cereal_data = pd.read_csv(url)
cereal_data.head()
```

```python
# Set the Seaborn style
sns.set(style="darkgrid")  # You can choose from "white", "dark", "whitegrid", "darkgrid"

# Calculate the average rating for each manufacturer
average_ratings = cereal_data.groupby('mfr')['rating'].mean().sort_values()

# Plot the ratings
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=average_ratings.index, y=average_ratings.values, palette='viridis')

# Set xticks range from 0-100
plt.xticks(rotation=0)  # Rotate labels for better readability
plt.xlim(-0.5, len(average_ratings)-0.5)
plt.ylim(0,100)

# Set labels and title
plt.xlabel('Manufacturer')
plt.ylabel('Rating')
plt.title('Rating of Cereals by Manufacturer')

# Show the plot
plt.tight_layout()
plt.show()
```
